define a position variable
compare number at current position
if smaller, return else increment till you exhaust every number


```python
def count_rotations(nums):
    position = 0
    while position < len(nums):
        if position > 0 and nums[position] < nums[position-1]:
            return position

        position+=1
    return 0
```

```python
def count_rotations_binary(nums):
    low, high = 0, len(nums)-1
    # low = nums[0] 
    # high = nums[-1]

    print('low:', low)
    print('high:', high)
    print('nums:', nums)
    


    while low <= high:
        middle_value = (low + high) // 2
        middle_number = nums[middle_value]
        print('middle_value:', middle_value,'middle_number:', middle_number, 'low:', low, 'high:', high )

        if middle_number> 0 and middle_number < middle_number - 1:           
            return middle_number
        elif middle_number < high:
            high = middle_number - 1
        elif middle_number > high:
            low = middle_number + 1

    return 0
```
