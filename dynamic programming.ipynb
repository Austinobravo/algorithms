```python
#test of only alphabets
test0 = {
    'input': {
        'seq1': 'sanitary',
        'seq2': 'semantics'
    },
    'output': 4
}


#test of numbers
test1 = {
    'input':{
        'seq1': [1,2,3,6,8,9,2,0],
        'seq2': [2,4,6,7,8,0]
    },
    'output': 4
}

#test of alphabets and numbers
test2={
    'input':{
        'seq1': [2,4,6,'d', 'g', 0],
        'seq2': [4,2,6,'g',5]
    },
    'output': 3
}

#test without subsequence
test3={
    'input':{
        'seq1': [2,4,6, 0],
        'seq2': [25,63,'g',5]
    },
    'output': 0
}
#test with same subsequence
test4={
    'input':{
        'seq1': [2,4,6, 0],
        'seq2': [2,4,6, 0]
    },
    'output': 4
}
#test thats empty
test5={
    'input':{
        'seq1': [],
        'seq2': []
    },
    'output': 0
}

#test that one is empty
test6={
    'input':{
        'seq1': [],
        'seq2': [2,4,5,'e']
    },
    'output': 0
}

```

```python
tests = [test0, test1,test2,test3,test4,test5,test6]
```

```python
def len_of_subsequence_dp(seq1,seq2):
    n1,n2 = len(seq1), len(seq2)
    table =[[0 for i in range(n2 +1)] for j in range(n1 +1)]
    for i in range(n1):
        for j in range(n2):
            if seq1[i] == seq2[j]:
                table[i+1][j+1] = 1 + table[i][j]
            else: 
                table[i+1][j+1] = max(table[i][j+1], table[i+1][j])
    return table[-1][-1]
```
